```python
import pandas as pd
import numpy as np
from IPython.display import display
# import matplotlib.pyplot as plt

df = pd.read_csv('credit_risk_data.csv')

print("Sample data:")

print(df.head())
df.info()
target_col = 'past_delinquencies'

# df = df.fillna(df.median(numeric_only=True))
# statistics = pd.DataFrame({
#     'Num_of_missing_values': df.isnull().sum(),
#     'Data_type': df.dtypes
# })
# print(statistics)
display(df.describe(include = "all").T)
```

```python
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

df = df.drop_duplicates()
print(f"Number of rows after removing duplicates: {len(df)}")

df.shape
```

```python
statistics = pd.DataFrame({
    'Num_of_missing_values': df.isnull().sum(),
    'Data_type': df.dtypes
})
print(statistics)
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        median_value = df[col].median()
        df[col] = df[col].fillna(median_value)
        print(f"Filled missing values in '{col}' with median value: {median_value}")
categorical_cols = df.select_dtypes(include=['str']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_value = df[col].mode()[0]
        df[col] = df[col].fillna(mode_value)
        print(f"Filled missing values in '{col}' with mode value: {mode_value}")
```

```python
df.to_csv('credit_risk_data_cleaned.csv', index=False)
```
